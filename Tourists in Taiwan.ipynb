Data Source:[政府資料開放平臺-歷年來台旅客統計](https://data.gov.tw/dataset/7322)
===

1956年到2016年來台旅客統計,旅客人數分為外籍及華僑兩類做紀錄

These table represents the number of visitors to Taiwan.
It is divided into two categories: foreign tourists and Chinese tourists.
    

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from matplotlib.font_manager import FontProperties
myfont = FontProperties(fname=r'/Library/Fonts/Songti.ttc', size=16) #設定中文字體
tourist_data = pd.read_csv('https://quality.data.gov.tw/dq_download_csv.php?nid=7322&md5_url=75b640397916e1d9ed2e308abbc0eb7b')
tourist_data.head(61)
```

## The picture shows the changes in the number of tourists to Taiwan over time
- **we find that there has sharply declined in 2003.
  The decline in the number of foreign tourists is the main reason.**
- **we also could see in 2004 and 2005, the number of Chinese tourists is heigher than foreign tourists.**

```python
#人數的datatype是str為了轉成int把逗號去除
for (i, row) in tourist_data.iterrows():
    tourist_data['總計人數'][i]=(int)(row['總計人數'].replace(',',''))
    tourist_data['外籍旅客人數'][i]=(int)(row['外籍旅客人數'].replace(',',''))
    tourist_data['華僑旅客人數'][i]=(int)(row['華僑旅客人數'].replace(',',''))
fig = plt.figure(figsize=(15, 7))
plt.xlim(1956, 2017)
plt.ylim(1e3, 12e6)
plt.xticks(np.arange(1956,2017,6),['{}年'.format(x) for x in np.arange(1956, 2017, 6)],fontproperties=myfont, fontsize=15)
plt.yticks(np.arange(1e3, 12e6, 1e6),np.arange(1e3, 12e6, 1e6))
'''
plt.plot(tourist_data['年別'], tourist_data['總計成長率'].apply(float), '-o')
plt.plot(tourist_data['年別'], tourist_data['外籍旅客成長率'].apply(float), '-o')
plt.plot(tourist_data['年別'], tourist_data['華僑旅客成長率'].apply(float), '-o')
'''
plt.plot(tourist_data['年別'], tourist_data['總計人數'], '-o',label="all tourists",color='#4E9A69')
plt.plot(tourist_data['年別'], tourist_data['外籍旅客人數'], 'r-o',label="foreign tourists",color='#DC87AF')
plt.plot(tourist_data['年別'], tourist_data['華僑旅客人數'], 'g-o',label="asian tourists",color='#6EC3D5')
plt.title("Number of visitors in Taiwan over the years(1956~2016)")
plt.xlabel('year',fontproperties=myfont)
plt.ylabel('tourist number',fontproperties=myfont)
plt.legend(labels=['Total','Foreign tourists', 'Chinese tourists'])
plt.show()
```

## The picture shows the percentage of forein tourists and overseas Chinese tourists to Taiwan and the changes of percentage over time.
- **we find that the percentage of overseas Chinese tourists gradually increasing in recent years.**

```python
fig = plt.figure(figsize=(15, 7))
plt.bar(tourist_data['年別'], round(tourist_data['外籍旅客占總計百分比']/100,2),
        color=['#F1B2C6'], width=0.9, align='center', edgecolor='none',label="foreign tourist")
plt.bar(tourist_data['年別'], round(tourist_data['華僑旅客占總計百分比']/100,2), bottom=round(tourist_data['外籍旅客占總計百分比']/100,2),
        color=['#A2DAF3'], width=0.9, align='center', edgecolor='none',label="asian tourist")
plt.title("Percentage of forein tourists and Chinese tourists to Taiwan over the years(1956~2016)",fontproperties=myfont)
plt.xlim(1956, 2017)
plt.ylim(0,1.01)
plt.xticks(np.arange(1956,2017,6),['{}年'.format(x) for x in np.arange(1956, 2017, 6)],fontproperties=myfont, fontsize=15)
plt.yticks(np.arange(0, 1.01, 0.25),
           ['{}%'.format(int(x * 100)) for x in np.arange(0, 1.01, 0.25)])
plt.ylabel('Percentage of tourist number')
plt.legend(labels=['Foreign tourist', 'Chinese tourist'],prop=myfont,loc='lower right')
```
